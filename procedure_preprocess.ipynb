```python
import pandas as pd

df_proc = pd.read_csv("procedures_icd.csv.gz", compression="gzip")

# Node file: one row per ICD procedure code + version
procedure_nodes = (
    df_proc[['icd_code', 'icd_version']]
    .drop_duplicates()
    .sort_values(['icd_version','icd_code'])
)

procedure_nodes.to_csv("procedure_nodes.csv", index=False)
procedure_nodes.head()
```

```python
import pandas as pd

df = pd.read_csv("procedures_icd.csv.gz", compression="gzip", dtype=str)

edges = (
    df[['hadm_id', 'icd_code', 'icd_version', 'chartdate', 'seq_num', 'subject_id']]
      .dropna(subset=['hadm_id','icd_code'])
      .drop_duplicates()
      .rename(columns={
          'hadm_id': ':START_ID(Admission)',
          'icd_code': ':END_ID(Procedure)'
      })
)

edges.to_csv("admission_has_procedure.csv", index=False)
print("✅ Rewrote admission_has_procedure.csv with neo4j-admin headers")
print(edges.head())
```

```python
import pandas as pd

df_proc_icd = pd.read_csv("d_icd_procedures.csv", dtype=str)

procedure_nodes = (
    df_proc_icd[['icd_code', 'icd_version', 'long_title']]
      .drop_duplicates()
      .rename(columns={
          'icd_code': 'icd_code:ID(Procedure)',
          'icd_version': 'icd_version',
          'long_title': 'long_title'
      })
)

procedure_nodes.to_csv("procedure_nodes.csv", index=False)

print("✅ procedure_nodes.csv created")
procedure_nodes.head()
```
